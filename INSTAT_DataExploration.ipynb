```python
import pandas as pd

# Define file paths and corresponding labels
files = {
    "data/instat/Games_GamesTotal_CS.xlsx": "GamesTotal",
    "data/instat/Games_EV_CS.xlsx": "EV",
    "data/instat/Games_PP_CS.xlsx": "PP",
    "data/instat/Games_PK_CS.xlsx": "PK"
}

# Initialize an empty list to store dataframes
games_list = []

# Loop through files, read, and append to the list
for file, game_type in files.items():
    df = pd.read_excel(file)
    df["Type"] = game_type  # Add a column indicating the source file type
    games_list.append(df)

# Concatenate all dataframes into one
games_df = pd.concat(games_list, ignore_index=True)

# Display the combined dataframe
print(games_df.head())

# Optionally, save the combined dataframe to a new Excel file
games_df.to_excel("data/instat/Combined_Games.xlsx", index=False)

# Combining Goalies data
goalies_files = {
    "data/instat/Goalies_GamesTotal_CS.xlsx": "GamesTotal",
    "data/instat/Goalies_EV_CS.xlsx": "EV",
    "data/instat/Goalies_PP_CS.xlsx": "PP",
    "data/instat/Goalies_PK_CS.xlsx": "PK"
}

goalies_list = []

for file, goalie_type in goalies_files.items():
    df = pd.read_excel(file)
    df["Type"] = goalie_type
    goalies_list.append(df)

goalies_df = pd.concat(goalies_list, ignore_index=True)

print(goalies_df.head())

goalies_df.to_excel("data/instat/Combined_Goalies.xlsx", index=False)

# Combining Skaters data
skaters_files = {
    "data/instat/Skaters_GamesTotal_CS.xlsx": "GamesTotal",
    "data/instat/Skaters_EV_CS.xlsx": "EV",
    "data/instat/Skaters_PP_CS.xlsx": "PP",
    "data/instat/Skaters_PK_CS.xlsx": "PK"
}

skaters_list = []

for file, skater_type in skaters_files.items():
    df = pd.read_excel(file)
    df["Type"] = skater_type
    skaters_list.append(df)

skaters_df = pd.concat(skaters_list, ignore_index=True)

print(skaters_df.head())

skaters_df.to_excel("data/instat/Combined_Skaters.xlsx", index=False)

# Combining Lines data

lines_files = {
    "data/instat/Lines_BasicLines_CS.xlsx": "Basic",
    "data/instat/Lines_DefencemenLines_CS.xlsx": "DF",
    "data/instat/Lines_ForwardLines_CS.xlsx": "FW",
    "data/instat/Lines_PenaltyKillersLines_CS.xlsx": "PK",
    "data/instat/Lines_PowerPlayersLines_CS.xlsx": "PP"
}

lines_list = []

for file, line_type in lines_files.items():
    df = pd.read_excel(file)
    df["Type"] = line_type
    lines_list.append(df)

lines_df = pd.concat(lines_list, ignore_index=True)

print(lines_df.head())

lines_df.to_excel("data/instat/Combined_Lines.xlsx", index=False)

```

```python
# ydata-profiling on all combined dataframes
from ydata_profiling import ProfileReport

profile_games = ProfileReport(games_df, title="Games Data Profiling Report")
profile_games.to_file("DataExploration/Games_Data_Profiling_Report.html")

profile_goalies = ProfileReport(goalies_df, title="Goalies Data Profiling Report")
profile_goalies.to_file("DataExploration/Goalies_Data_Profiling_Report.html")

profile_skaters = ProfileReport(skaters_df, title="Skaters Data Profiling Report")
profile_skaters.to_file("DataExploration/Skaters_Data_Profiling_Report.html")

profile_lines = ProfileReport(lines_df, title="Lines Data Profiling Report")
profile_lines.to_file("DataExploration/Lines_Data_Profiling_Report.html")
```

```python
# Define a function to rank lines based on a specific column
def rank_lines(df, situation, column, ascending=False):
    # Filter the dataframe for the specific situation
    filtered_df = df[df['Type'] == situation]
    # Rank the lines based on the specified column
    ranked_df = filtered_df.sort_values(by=column, ascending=ascending)
    return ranked_df

# Rank lines for Even Strength (EV)
ranked_lines_ev = rank_lines(lines_df, 'EV', 'CORSI+')
print("Top 5 Even Strength Lines based on CORSI+:\n", ranked_lines_ev.head())

# Rank lines for Power Play (PP)
ranked_lines_pp = rank_lines(lines_df, 'PP', 'Goals')
print("\nTop 5 Power Play Lines based on Goals:\n", ranked_lines_pp.head())

# Rank lines for Penalty Kill (PK)
ranked_lines_pk = rank_lines(lines_df, 'PK', 'Shots on goal against', ascending=True)
print("\nTop 5 Penalty Kill Lines based on least Shots on goal against:\n", ranked_lines_pk.head())

# Rank defensemen lines
ranked_lines_d = rank_lines(lines_df, 'D', 'CORSI+')
print("\nTop 5 Defensemen Lines based on CORSI+:\n", ranked_lines_d.head())

# Rank forwards lines
ranked_lines_f = rank_lines(lines_df, 'F', 'CORSI+')
print("\nTop 5 Forwards Lines based on CORSI+:\n", ranked_lines_f.head())

# Rank Basic Lines
ranked_lines_basic = rank_lines(lines_df, 'Basic', 'CORSI+')
print("\nTop 5 Basic Lines based on CORSI+:\n", ranked_lines_basic.head())
```
