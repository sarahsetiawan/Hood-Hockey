```python
import pandas as pd

# Load the Excel file
skaters_df = pd.read_excel('/Users/sarahsetiawan/Desktop/CSCAPSTONE/Hood-Hockey/data/instat/Combined_Skaters.xlsx')

# Display the first few rows of the dataframe
skaters_df.head()
```

```python
skaters_df.replace('-', 0, inplace=True)
skaters_df.head()
```

```python
percentage_columns = ['Faceoffs won, %', 'Faceoffs won in DZ, %', 'Faceoffs won in NZ, %', 'Faceoffs won in OZ, %']
for col in percentage_columns:
    skaters_df[col] = skaters_df[col].str.replace('%', '').astype(float)
```

```python
# Convert '0' values to '00:00' in specified time columns if they exist
time_columns = ['Time on ice']
for col in time_columns:
    if col in skaters_df.columns:
        skaters_df[col] = skaters_df[col].replace('0', '00:00')
        # Fill NaN values with '00:00' before splitting
        skaters_df[col] = skaters_df[col].fillna('00:00')
        # Split the time into minutes and seconds
        skaters_df[[f'{col} (mins)', f'{col} (secs)']] = skaters_df[col].str.split(':', expand=True)
        # Fill NaN values in the new columns with 0 and convert to integers
        skaters_df[f'{col} (mins)'] = skaters_df[f'{col} (mins)'].fillna(0).astype(int)
        skaters_df[f'{col} (secs)'] = skaters_df[f'{col} (secs)'].fillna(0).astype(int)
    else:
        print(f"Column '{col}' not found in DataFrame")

# Display the first few rows of the dataframe
skaters_df.head()
```

```python
skaters_df.to_excel('/Users/sarahsetiawan/Desktop/CSCAPSTONE/Hood-Hockey/DataCleaning/skaters_cleaned.xlsx', index=False)
```
